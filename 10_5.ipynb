```python
import numpy as np
import matplotlib.pyplot as plt
from pywaffle import Waffle as wf
import pandas as pd
from wordcloud import WordCloud,STOPWORDS

```

```python
x=range(1,6)
y1=[1,4,6,8,9]
y2=[2,2,7,10,12]
y3=[2,8,5,10,6]
plt.stackplot(x,y1,y2,y3,labels=["A","B","C"],colors=["yellow","red","black"])
plt.legend(loc='upper left')
plt.show()
```

```python
stopwords=set(STOPWORDS)
stopwords
```

```python
with open("alice_in_wonderland.txt")as f:
    alice_novel=f.read()
alice_wc=WordCloud(background_color="lightpink",max_words=2000,stopwords=stopwords)
alice_wc.generate(alice_novel)
plt.imshow(alice_wc)

plt.axis('off')
plt.show()  
    
```

```python
stopwords.add("said")
alice_wc.generate(alice_novel)
fig=plt.figure(figsize=(14,16))
plt.imshow(alice_wc)

plt.axis('off')
plt.show()  
    
```

```python
import seaborn as sns
dataset=pd.read_csv('auto-mpg.csv')
sns.regplot(x="mpg",y="acceleration",data=dataset)
plt.show()
```

```python
ax=sns.regplot(x="weight",y="displacement",data=dataset,marker="*")
ax.set(xlabel="weigth",ylabel="displacement")
plt.figure(figsize=(14,16))
sns.set(font_scale=1.8)

ax.set_title("Weigth v/s Displacement")
plt.show()
```
