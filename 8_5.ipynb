```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def find_outliers(ds,col):
    quart1=ds[col].quantile(0.25)
    quart3=ds[col].quantile(0.75)
    IQR=quart3-quart1
    low_val=quart1-(1.5*IQR)    
    high_val=quart3+(1.5*IQR)
    ds=ds.loc[(ds[col]>low_val) & (ds[col]<high_val)]
    return ds
```

```python
dataset=pd.read_csv("auto-mpg.csv")
find_outliers(dataset,'acceleration')
```

```python

# data regarding income and expenses are given show the graph which will fill the area with green colour when income grater than expanses and when income<=expenses then fill the area with red color
time=np.arange(10)
income=np.array([5,9,6,6,10,7,6,4,4,5])
expenses=np.array([6,6,8,3,6,9,7,8,6,6])
fig,ax=plt.subplots(figsize=(10,10))
ax.plot(time,income,color='green')
ax.plot(time,expenses,color='red')

ax.fill_between(time,income,expenses,where=(income>expenses),color="green",alpha=0.25,label="positive",interpolate=True)
ax.fill_between(time,income,expenses,where=(income<=expenses),color="red",alpha=0.25,label="negative",interpolate=True)
plt.xlabel('time')
plt.ylabel("income vs expenses comparision ")
ax.legend()
plt.show()


```
